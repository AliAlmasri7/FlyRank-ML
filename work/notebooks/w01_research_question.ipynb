## Research Question

**Decision**

Which pages should be prioritized for SEO investigation based on their observed ranking signals?

**Action**

An SEO specialist reviews the highest-priority pages, updates content, improves on-page optimization, or performs additional analysis before rankings decline further.

**Cost of a Wrong Recommendation**

If the system recommends the wrong pages, valuable time may be spent reviewing healthy pages while pages that actually need attention remain unchanged. This reduces the efficiency of SEO efforts and may lead to missed opportunities for improving search performance.

**Why ML or Data Analysis Helps**

Many ranking signals interact with each other. Instead of relying on a single manually defined rule, data analysis and machine learning can identify combinations of signals that are more strongly associated with declining performance and help prioritize pages more consistently.

```python
#Load the dataset
import os, sys, subprocess

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", "requirements.txt"], check=True)
elif os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("..")

import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# The label: a page is 'declining' when its recent trend is down. Simple, honest starter label.
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)
print(df.shape[0], "pages |  declining rate:", round(df["is_declining_label"].mean(), 3))
```

```python
df.head()
```

```python
#Pick ranking signals to show the results
signals = [
    "avg_position",
    "ctr",
    "impressions_90d",
    "days_since_last_update"
]

df.groupby("is_declining_label")[signals].mean().round(2)
```

An initial look at the starter dataset shows measurable differences between declining and non-declining pages. Pages labeled as declining have a much lower average CTR (0.32 vs. 0.73), slightly fewer impressions over the last 90 days (4919 vs. 5533), and have gone longer without being updated on average (49.25 vs. 42.37 days). Although the average position is similar (15.94 vs. 16.82), these observed differences suggest that multiple ranking signals may be associated with page decline. This makes Ranking Signal Analysis a suitable direction for investigating which signals are most useful for prioritizing pages for SEO review.
